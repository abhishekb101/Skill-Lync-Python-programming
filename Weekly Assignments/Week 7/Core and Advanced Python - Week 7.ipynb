**Q1** - *Create a class called number which contains an integer. It should contain various methods to set the value of the integer, return the integer value, print the integer value, check whether the integer is negative.*

```python
class number:
    
    #Initialising the attributes
    def __init__(self):
        self._val = int(input('Enter an integer: '))
    
    #Returning the value of the integer 
    def showVal(self):
        return self._val
    
    #Printing the value of the integer
    def printVal(self):
        print('Printing the integer value: {}'.format(self._val))
    
    #Checking whether it is negative or not
    def checkInt(self):
        if self._val < 0:
            return '{} is negative'.format(self._val)
        else:
            return '{} is non negative'.format(self._val)
        
#Creating an object of the class
num = number()

print('\nRunning different methods on the object num:')
print('Returning the integer value: {}'.format(num.showVal()))
num.printVal()
print(num.checkInt())
```

**Q2** - *Create a class that can calculate the perimeter/ circumference and area of a regular shape. The class should also have a provision to accept the data relevant to the shape.*

```python
#Mapping basic shape_type -- {rectangle: 1, circle = 2, square = 3}
import math

class shape:
    
    #Initialising the shape_type for common shapes
    def __init__(self):
        self._shape_type = int(input('Enter the shape_type: '))
    
    #Calculating the perimter of the rectangle
    def perimeter_calc(self):
        #Calculating perimeter of a rectangle
        if self._shape_type == 1:
            self._length = float(input('Enter the length of the rectangle: '))
            self._breadth = float(input('Enter the length of the rectangle: '))
            return 2 * (self._length + self._breadth)
        
        #Calculating circumference of a circle
        elif self._shape_type == 2:
            self._radius = float(input('Enter the radius of the circle: '))
            return 2 * math.pi * self._radius
        
        #Perimeter of the square
        elif self._shape_type == 3:
            self._side = float(input('Enter the side of a square: '))
            return 4 * self._side
    
    #Calculating the area of the rectangle
    def area_calc(self):
        #Calculating area of a rectangle
        if self._shape_type == 1:
            return self._length * self._breadth
        
        #Calculating area of a circle
        elif self._shape_type == 2:
            return math.pi * (self._radius ** 2)
        
        #Area of a square
        elif self._shape_type == 3:
            return self._side ** 2
                
r = shape()

print('\nThe perimter of the rectangle: {:.2f}'.format(r.perimeter_calc()))
print('The area of the rectangle: {:.2f}'.format(r.area_calc()))
```

**Q4** - *Modify the student class defined in Q. No. 3 in so that it now includes a class variable to count the number of student objects created from this class and also add a class function in order to display the value of the class variable. The modified class should be named as student1 class.*

```python
#Assuming the question talks about the student class that has been created in Q3.
#As Q1 contains the number class. Hence, creating modified student class which builds on Q3

class student1:
    
    #Class variable created to count total no. of student objects created
    instance_cnt = 0
    
    #Initialising the class
    def __init__(self, rollno, sname, branch):
        #Invoking the class method
        student1.increment_count()
        
        self._rollno = rollno
        self._sname = sname
        self._branch = branch
        
    #Displaying data members
    def display_data(self):
        print('The roll no of the student: {}'.format(self._rollno))
        print('The surname of the student: {}'.format(self._sname))
        print('The branch of the student: {}'.format(self._branch))
    
    #Defining a class function to aggregate the total num of students added
    @classmethod
    def increment_count(cls):
        
        #Incrementing the value of the class variable
        cls.instance_cnt += 1
    
    def __del__(self):
        print('Deleting the object {}'.format(str(self)))

#Creating instances of the class student and running the methods on them
s1 = student1(100, 'Bhattacharyya', 'Mechatronics')
s1.display_data()

print('\n' + '*'*10 + '\n')

s2 = student1(101, 'Sriram', 'Electronics')
s2.display_data()

print('\n' + '*'*10 + '\n')

s3 = student1(102, 'Patel', 'Mechanical')
s3.display_data()

print('\nThe number of student objects created: {}'.format(student1.instance_cnt))
```

**Q6** - *Create a class called Date. It should contain data members day, month and year. It should also contain functions to set and display date, as well as function to initialize date objects. Overload the == operator to check if two dates are equal*

```python
#Assuming Dates are of form -- 5 January 2000 (int, str, int)

class Date:
    
    #Initialising the data members of the class
    def __init__(self, day=0, month='', year=0):
        self._day = day
        self._month = month
        self._year = year
    
    #Creating a method for setting dates
    def set_date(self, day, month, year):
        self._day = day
        self._month = month
        self._year = year
        
    #Creating a method to print the date in a standardized format    
    def display_date(self):
        print('The day: {}'.format(self._day))
        print('The month: {}'.format(self._month))
        print('The year: {}'.format(self._year))
             
    def __str__(self):
        result = ' '.join(list(map(str, (self._day, self._month, self._year))))
        return result
        
    #Creating a == operator overload to compare two objects of the date class
    def __eq__(self, other):
        
        #If all the three values match between two dates then they are equal - printing the same
        if all([self._day == other._day, self._month == other._month, self._year == other._year]):
            print('The two dates are equal.')
        else:
            print("The two dates aren't equal.")
    
#Creating two objects of the class date for running its methods

d1 = Date()
d2 = Date()

#Setting the two dates
d1.set_date(5, 'February', 2021)
d2.set_date(5, 'March', 2021)

#Displaying the two dates
print('d1:')
d1.display_date()
print('\nd2:')
d2.display_date()

print('\n' + '*' * 10 + '\n')

#Printing the two dates
print('The date d1: {}'.format(d1))
print('The date d2: {}'.format(d2))

#Comparing the two dates
print('\n' + '*' * 10 + '\n')
d1 == d2
```

**Q7** - *Create a class called weather that has a list containing weather parameters. Overload the in operator that checks whether an item is present in the list or not.*

```python
#Assuming that the list contans weather parameters and not its value
#As each of them have different units of measurement

class weather:
    #Initialise the list as the data member
    def __init__(self):
        self._weather_param = input('Enter weather parameters in a space seperated format: ').split(' ')
    
    #Overridding the in operator using the __contains__ constructor 
    def __contains__(self, element):
        return element in self._weather_param
    
    #Printing the various weather parameters
    def __str__(self):
        return '\nThe weather list contains parameters: {}\n'.format(self._weather_param)

#Creating an instance of the class weather
w1 = weather()  
print(w1)

#Checking whether an item is present in list or not
param = 'wind'
if param in w1:
    print('{} is present in the weather parameter list.'.format(param))
else:
    print('{} is not present in the weather parameter list.'.format(param))
```

**Q8** - *Create a class called complex containing real and imaginary parts and then use it to check whether two objects are of the same type, whether their attributes are same and whether they are pointing to the same object*

```python
class Complex:
    
    #Initialising the real and imaginary parts of the complex number
    def __init__(self, real, imaginary):
        self._re = real
        self._im = imaginary
    
    #Printing the complex number
    def __str__(self):
        return '{}{:+}j'.format(self._re, self._im)

    #Overloading the equal to operator to check whether attributes of both are same or not    
    def __eq__(self, other):
        if (self._re == other._re) and (self._im == other._im):
            print('Their attributes are same.')
        else:
            print('Their attributes are different.')

#Creating two instances of the class             
c1 = Complex(3, 5)
c2 = Complex(3, 5)

#Printing the two instances of the class Complex
print('The first complex no: {}'.format(c1))
print('The second complex no: {}'.format(c2))

print('\n' + '*' * 10 + '\n')

#Checking whether both objects are of same time using isinstance
if isinstance(c1, Complex) and isinstance(c2, Complex):
    print('Both objects are of the same type.')
else:
    print('Both objects are not of the same type.')
    
#Checking whether their attributes are same using __eq__ operator overloading
c1 == c2

#Checking whether they are pointing to the same object or not -- Memory location
#Even if there attributes are same but when initialised separately they will never point to same object

if id(c1) == id(c2) and c1 is c2:
    print('They point to the same object.')
else:
    print("They don't point to the same object.")    
```

**Q9** - *Create a class called department with the attributes deptname and deptid. Modify the student class defined in Q.No. 3 such that it contains an object of department class. Name the new modified student class as student2.*

```python
class department:
    
    #Initialising the attributes 
    def __init__(self, deptname = '', deptid = ''):
        self._deptname = deptname
        self._deptid = deptid
        
    #Setting the attributes of department from user entry
    def set_data(self):
        self._deptname = input('Enter a department name: ')
        self._deptid = input('Enter a department ID: ')

    #Displaying the department data
    def display_data(self):
        print('The department name: {}'.format(self._deptname))
        print('The department id: {}'.format(self._deptid))

class student2:
    
    #Initialising the class
    def __init__(self, rollno, sname, branch):
        self._rollno = rollno
        self._sname = sname
        self._branch = branch
        
        #Creating a container and setting params -- self._dept is the object of the department class inside student2
        self._dept = department()
        self._dept.set_data()
        
    #Displaying data members
    def display_data(self):
        print('\nThe roll no of the student: {}'.format(self._rollno))
        print('The surname of the student: {}'.format(self._sname))
        print('The branch of the student: {}'.format(self._branch))
        self._dept.display_data()
        
    def __del__(self):
        print('Deleting the object {}'.format(str(self)))

#Creating instances of the class student2 and running the methods on them -- In turn calls department class
s1 = student2(100, 'Bhattacharyya', 'Engineering')
s1.display_data()

print('\n' + '*'*10 + '\n')

s2 = student2(101, 'Sriram', 'Engineering')
s2.display_data()    
```

**Q10** - *Create a class called Person. From this class, inherit another class called student. Assume suitable data members and member functions for these two classes. In both the classes, define the __str__ function and demonstrate method overriding.*

```python
class Person:
    
    #Initialising the Person class 
    def __init__(self, fname, lname, age):
        self._fname = fname
        self._lname = lname
        self._age = age
    
    #Checking whether the person is an adult or not
    def adult_check(self):
        if self._age >= 18:
            print('{} is an adult.'.format(self._fname))
        else:
            print('{} is not an adult.'.format(self._fname))
    
    #Printing the instance of the class
    def __str__(self):
        return '{} {} is {} years old.'.format(self._fname, self._lname, self._age)
    
    
#Defining a base class Student which inherits from person
#Using the above defined functionality itself

class Student(Person):
    
    #Initialising the class and overridding the __init__ constructor of the parent class
    def __init__(self, fname, lname, age, rollno, dept):
        
        #Initialising the parent class
        super().__init__(fname, lname, age)
        self._rollno = rollno
        self._dept = dept
        
    #Displaying data members
    def display_data(self):
        print('The first name of the student: {}'.format(self._fname))
        print('The last name of the student: {}'.format(self._lname))
        print('The age of the student: {}'.format(self._age))
        print('The roll no of the student: {}'.format(self._rollno))
        print('The department of the student: {}'.format(self._dept))
    
    #Overidding the parent class __str__ constructor -- Method overridding
    def __str__(self):
        return 'Roll no: {}, {} {} is {} years old and is from the {} department'.format(self._rollno, self._fname,
                                                                                self._lname, self._age, self._dept)
    
#Demonstrating the methods of the two classes 
p1 = Person('Sagarika', 'Sriram', 24)
p2 = Person('Karim', 'Khoja', 26)

#Running the adult check method on the two instances
p1.adult_check()
p2.adult_check()

#Printing the instances
print('\n' + '*'*10)
print('Calling the __str__ constructor of the parent class Person:\n')
print(p1)
print(p2)

#Creating instances of the class student
s1 = Student('Abhishek', 'Bhattacharyya', 24, 100, 'Mechatronics')
s2 = Student('Roshnee', 'Patel', 26, 101, 'Electronics')

#Displaying the data
print('\n' + '*'*10)
print('Displaying the two instances of the class Student:\n')
s1.display_data()
print('\n')
s2.display_data()

#As student inherits from Person but has its own __str__ constructor 
#It will override the parent classes __str__ method. Evident below
print('\n' + '*'*10)
print('Calling the __str__ constructor of the child class Student:- Method overridding\n')
print(s1)
print(s2)
```
