**Q4** - *Create a tkinter application to accept radius of a circle and display the area.*

```python
#Creating a window 'root' and adding a title to it

root = Tk()
root.title('Displaying area of the circle')

#Adding a frame inside the root window 
frame = LabelFrame(root, padx = 20, pady = 20)
frame.pack(padx = 10, pady = 10)

#Creating a label widget to let user know where to enter radius
l1 = Label(frame, text = 'Enter radius of the circle: ', padx = 3, pady = 3, width = 18, justify = 'left',
          font=('calibre',12, 'bold'))

#Creating an entry widget to accept the value of the radius
e1 = Entry(frame, width = 12, borderwidth = 3)

#Creating label widget to display the area calculated
l2 = Label(frame, text = 'Area of the circle: ', padx = 3, pady = 3, width = 15, justify = 'left',
          font=('calibre',12, 'bold'))

#Creating an entry widget to display the value of the area calculated
e2 = Entry(frame, width = 12, borderwidth = 3)

#Defining a function to calculate area
def area():
    """
    Calculates the area using the input radius value 
    Rounded upto 3 decimal places
    """
    out = round(math.pi * (float(e1.get()) ** 2), 3)
    
    #Removing any number entered already in the area cell
    e2.delete(0, END)
    
    #Inserting the value of the output
    e2.insert(0, out)
    
def clear_screen():
    """
    Clears the entry widgets to start the calculation again
    """
    e1.delete(0, END)
    e2.delete(0, END)
    
#Creating a button to actuate the conversion and display results
b1 = Button(frame, text = 'Calculate area', padx = 3, pady = 3, width = 9, command = area)

#Inseting a clear button to rest entries by user
b2 = Button(frame, text = 'Clear', padx = 3, pady = 3, command = clear_screen)

#Placing all widgets using grid in the root window
l1.grid(row = 0, column = 0)
e1.grid(row = 0, column = 1, padx = 3, pady = 3)
l2.grid(row = 1, column = 0)
e2.grid(row = 1, column = 1, padx = 3, pady = 3)
b1.grid(row = 2, column = 1)
b2.grid(row = 2, column = 0)

#To allow the program to iterate continously unless closed by the user
root.mainloop()
```

**Q5** - *Create a tkinter application to accept temperature in Celsius and convert and display the temperature in Fahrenheit.*

```python
#Creating a window 'root' and adding a title to it

root = Tk()
root.title('Converting temperature in both C & F')

#Adding a frame inside the root window 
frame = LabelFrame(root, padx = 20, pady = 20)
frame.pack(padx = 10, pady = 10)

#Creating a label widget to let user know where to enter temp in celsius
l1 = Label(frame, text = 'Temperature in Celsius: ', padx = 3, pady = 3, width = 18, justify = 'left',
          font=('calibre',12, 'bold'))

#Creating an entry widget to accept the value of the temperature in celsius
e1 = Entry(frame, width = 12, borderwidth = 3)

#Creating label widget to display the temperature in farhenheit
l2 = Label(frame, text = 'Temperature in Fahrenheit: ', padx = 3, pady = 3, width = 21, justify = 'left',
          font=('calibre',12, 'bold'))

#Creating an entry widget to display the value of the temp in F
e2 = Entry(frame, width = 12, borderwidth = 3)

#Defining a function to convert temperature from C to F and vice versa 
def convert_tmp():
    """
    Converts the temperature from celsius to farhenheit and
    vice versa as well using if-else statement
    """
    #Checking to see whether the user has entered a temp value or not in first field
    if len(e1.get()) > 0:
        
        #Converting to fahrenheit and inserting the value in the second field
        tmp_f = ((9/5) * float(e1.get())) + 32
        e2.delete(0, END)
        e2.insert(0, tmp_f)
    
    #Checking whether the user has entered a temp value in the second field
    elif len(e2.get()) > 0:
        
        #Converting to celsius and inserting the value in the first field
        tmp_c = (5/9) * (float(e2.get()) - 32)
        e1.delete(0, END)
        e1.insert(0, tmp_c)
    
def clear_screen():
    """
    Clears the entry widgets to start the conversion again
    """
    e1.delete(0, END)
    e2.delete(0, END)
    
#Creating a button to actuate the conversion and display results
b1 = Button(frame, text = 'Convert', padx = 3, pady = 3, width = 9, command = convert_tmp)

#Inseting a clear button to rest entries by user
b2 = Button(frame, text = 'Clear', padx = 3, pady = 3, command = clear_screen)

#Placing all widgets using grid in the root window
l1.grid(row = 0, column = 0)
e1.grid(row = 0, column = 1, padx = 3, pady = 3)
l2.grid(row = 1, column = 0)
e2.grid(row = 1, column = 1, padx = 3, pady = 3)
b1.grid(row = 2, column = 1)
b2.grid(row = 2, column = 0)

#To allow the program to iterate continously unless closed by the user
root.mainloop()
```

**Q6** - *Create a tkinter application to accept distance in kilometers and convert and display distance in miles.*

```python
#Creating a window 'root' and adding a title to it

root = Tk()
root.title('Converting distance in kms to miles')

#Adding a frame inside the root window 
frame = LabelFrame(root, padx = 20, pady = 20)
frame.pack(padx = 10, pady = 10)

#Creating a label widget to let user know where to enter distance in kms
l1 = Label(frame, text = 'Distance (kms): ', padx = 3, pady = 3, width = 14, justify = 'left',
          font=('calibre',12, 'bold'))

#Creating an entry widget to accept the value of distance in kms
e1 = Entry(frame, width = 12, borderwidth = 3)

#Creating label widget to display the distance in miles
l2 = Label(frame, text = 'Distance (miles): ', padx = 3, pady = 3, width = 14, justify = 'left',
          font=('calibre',12, 'bold'))

#Creating an entry widget to display the value of distance in miles
e2 = Entry(frame, width = 12, borderwidth = 3)

#Defining a function to convert distance in miles
def convert_kms():
    """
    Converts the distance in kms to miles upon actuation of button
    """     
    #Converting to miles and inserting the value in the second field
    miles = round(float(e1.get())/1.609, 3)
    e2.delete(0, END)
    e2.insert(0, miles)
    
def clear_screen():
    """
    Clears the entry widgets to start the calculation again
    """
    e1.delete(0, END)
    e2.delete(0, END)
    
#Creating a button to actuate the conversion and display results
b1 = Button(frame, text = 'Convert', padx = 3, pady = 3, width = 9, command = convert_kms)

#Inseting a clear button to rest entries by user
b2 = Button(frame, text = 'Clear', padx = 3, pady = 3, command = clear_screen)

#Placing all widgets using grid in the root window
l1.grid(row = 0, column = 0)
e1.grid(row = 0, column = 1, padx = 3, pady = 3)
l2.grid(row = 1, column = 0)
e2.grid(row = 1, column = 1, padx = 3, pady = 3)
b1.grid(row = 2, column = 1)
b2.grid(row = 2, column = 0)

#To allow the program to iterate continously unless closed by the user
root.mainloop()
```

**Q7** - *Create a tkinter application to display a menu with three items. A separate message should be display when each of the three menu items is clicked.*

```python
#Creating a window root, adding a title to it and setting its geometry

root = Tk()
root.title("Working with Menu's")
root.geometry("250x100")

#Creating a top level menu and configuring it 
menubar = Menu(root)
root.config(menu = menubar)

#Creating a pull down menu and adding its label
names = Menu(menubar)
menubar.add_cascade(label = 'People', menu = names)

#Defining a message to display when menu items are pressed
def message1():
    """
    Displaying a message when the menu item 'Abhishek' is clicked
    """
    #Setting the label text
    l.config(text = 'Abhishek is 24 years old.')
    
def message2():
    """
    Displaying a message when the menu item 'Sagarika' is clicked
    """
    #Setting the label text
    l.config(text = 'Sagarika is 23 years old.')
    
def message3():
    """
    Displaying a message when the menu item 'Roshnee' is clicked
    """
    #Setting the label text
    l.config(text = 'Roshnee is 26 years old.')
    
#Adding items under the first tab 'People'
names.add_command(label = 'Abhishek', command = message1)
names.add_separator()
names.add_command(label = 'Sagarika', command = message2)
names.add_separator()
names.add_command(label = 'Roshnee', command = message3)

#Creating a label widget and packing it when the menu item is pressed
l = Label(root, padx = 10, pady = 10, font = ('calibre',12, 'bold'))
l.pack()
    
#To allow the program to iterate continously unless closed by the user
root.mainloop()
```

**Q8** - *Create a tkinter application to display a radio button group with three items. A separate message should be display when each of the three radio buttons items is clicked.*

```python
#Creating a window root and adding a title to it

root = Tk()
root.title('Working with Radio buttons')

#Adding a frame inside the root window 
frame = LabelFrame(root, text = 'Favourite sports..', padx = 60, pady = 10)
frame.pack(padx = 10, pady = 15)

#Creating a list of tuples 'sports' to store the display text and value of each radio button
sports = [('Football', 'first'), ('Tennis', 'second'), ('Cricket', 'third')] 

#Defining a variable s from which we can obtain the current value of the radiobutton
s = StringVar()

#Setting the default option as first -- Football will be initially selected when window appears
s.set('first')

def click(value):
    """
    Displaying a message as label when any of the radio buttons are pressed by the user
    """
    #Conditional logic to display separate message depending on radio button selected
    if value == 'first':
        
        #Configuring the label widget and displaying the message based on 's.get()'
        message = 'Football is my {} favourite sport.'.format(value)
        l.config(text = message)
        
    elif value == 'second':
        #Configuring the label widget and displaying the message based on 's.get()'
        message = 'Tennis is my {} favourite sport.'.format(value)
        l.config(text = message)
        
    else:
        #Configuring the label widget and displaying the message based on 's.get()'
        message = 'Cricket is my {} favourite sport.'.format(value)
        l.config(text = message)

#Iterating over sports to create 3 radiobuttons
for sport, rank in sports:
    #Creating a radio button widget and upon pressing it displays the message 
    Radiobutton(frame, text = sport, variable = s, value = rank, command = lambda: click(s.get())).pack(anchor = 'w')
    
#Defining a label widget to display the mesaage and packing it
l = Label(frame, padx = 3, pady = 3, font = ('calibre',12, 'bold'))
l.pack()

#To allow the program to iterate continously unless closed by the user
root.mainloop()
```

**Q9** - *Create a tkinter application to display three check boxes. A separate message should be display when each of the three check boxes items is selected.*

```python
#Creating a window root and adding a title to it

root = Tk()
root.title('Working with Check buttons')

#Adding a frame inside the root window 
frame = LabelFrame(root, text = 'Hobbies..', padx = 60, pady = 10)
frame.pack(padx = 10, pady = 15)

#Creating a list of 'hobbies' to store the display text of each check button
hobbies = ['Football', 'Cooking', 'Reading self-dev novels'] 

#Defining variables from which we can obtain the current value of the check button
h1, h2, h3 = StringVar(), StringVar(),  StringVar()

#Setting the check buttons initially turned off at start of application
h1.set('off')
h2.set('off')
h3.set('off')

def click1():
    """
    Displaying a message as label when the check buttons are selected by the user
    and displaying empty string when deselected
    """
    #Checking whether the first check button is selected
    if h1.get() == 'on':
        
        #Setting the label 'l1' text paramter
        l1.config(text = 'Football is one of my Hobbies.')
    else:
        #Setting label text as empty when first check button is de-selected
        l1.config(text = '')
    
def click2():
    """
    Displaying a message as label when the check buttons are selected by the user
    and displaying empty string when deselected
    """
    #Checking whether the second check button is selected
    if h2.get() == 'on':
        
        #Setting the label 'l2' text paramter
        l2.config(text = 'Cooking is one of my Hobbies.')
    else:
        #Setting label text as empty when second check button is de-selected
        l2.config(text = '')
        
def click3():
    """
    Displaying a message as label when the check buttons are selected by the user
    and displaying empty string when deselected
    """
    #Checking whether the first check button is selected
    if h3.get() == 'on':
        
        #Setting the label 'l3' text paramter
        l3.config(text = 'Reading novels is one of my Hobbies.')
    else:
        #Setting label text as empty when third check button is de-selected 
        l3.config(text = '')
    
#Creating checkbuttons, setting all its necessary parameters and packing the widgets
Checkbutton(frame, text = hobbies[0], variable = h1, onvalue = 'on',
            offvalue = 'off', command = click1).pack(anchor = 'w')
Checkbutton(frame, text = hobbies[1], variable = h2, onvalue = 'on',
                 offvalue = 'off', command = click2).pack(anchor = 'w')
Checkbutton(frame, text = hobbies[2], variable = h3, onvalue = 'on',
                 offvalue = 'off', command = click3).pack(anchor = 'w')

#Defining a label widget to display the mesaage and packing it
l1 = Label(frame, padx = 5, pady = 5, font = ('calibre',12, 'bold'))
l2 = Label(frame, padx = 5, pady = 5, font = ('calibre',12, 'bold'))
l3 = Label(frame, padx = 5, pady = 5, font = ('calibre',12, 'bold'))

#Packing the labels
l1.pack()
l2.pack()
l3.pack()

#To allow the program to iterate continously unless closed by the user
root.mainloop()
```
