**Q3** - *Write a program that accepts n positive integers as input and prints their square. If a negative number is entered, then raise ValueError exception and display a relevant error message and make an exit.*

```python
#Solving this by creating an iterable object

class square:
    
    #Initialising the variables required
    def __init__(self, n):
        self._n = n
        
        #Starting the loop from 1
        self._val = 1
    
    #Making the object iterable    
    def __iter__(self):
        return self
    
    #Defining the square function within the class
    def sq_num(self, x):
        """
        Takes a number and returns its square
        """
        return x**2
    
    #Logic of returning the next element in the sequence
    def __next__(self):
        
        #Stopping the iteration if loop reaches n
        if self._val > self._n:
            raise StopIteration('\nThe loop has been completed !!')
        
        else:
            out = self.sq_num(self._val)
            
            #Incrementing the val by 1 on each iteration
            self._val += 1
            
            #Returning the squared value as output as ouput
            return out

#Entering the input num value
num = int(input('Enter a number: '))
    
if num < 0:
    raise ValueError('Entered a negative number to iterate over. Only positive integers acceptable as input.')

print('\nThe square of the first {} positive numbers:'.format(num))
#Iterating over the square iterable class if input is positive integer
for ele in square(num):
    print(ele, end = ' ')
```

**Q4** - *Write a program that receives an integer as input and finds its factorial. If a non-integer input is entered, then report an error and accept the input again. Continue this process until correct input is entered.*

```python
#Assuming only positive integers are valid forms of input. 
#As negative factorials are complex numbers which can be avoided

#Defining the factorial function in O(n) time complexity and O(1) space complexity
def fact(num):
    """
    Takes a positive integer as input and returns
    the factorial of the same
    """
    prod = 1
    
    #Iterating from 1 to n
    for i in range(1, num+1):
        prod *= i
        
    return prod

#Using while loop to loop until the right program is entered
while True:
    
    try:
        #Accepting the user input only when its a positive integer
        num = int(input('Enter an integer: '))
        
        #Raising an error when a negative integer is entered 
        if (num < 0):
            raise ValueError
        
        print('\nThe factorial of the number {}: {}'.format(num, fact(num)))
        break
        
    except ValueError:
        print('\nIncorrect input value. Kindly enter again')
        continue
```

**Q6** - *Write a program that implements iterator class called Prime and which prints prime numbers up to n.*

```python
class Prime:
        
    #Initialising the variables required
    def __init__(self, n):
        self._n = n
        
        #Starting the loop from 2 -- Smallest prime number
        self._val = 2
    
    #Making the object iterable    
    def __iter__(self):
        return self

    #Defining a function to check whether the value is prime or not
    def is_prime(self, num1):
        """
        Returns a boolean output on whether num1
        is prime or not
        """
        
        #Creating local temp variables for displaying result
        flag = True
        idx = 2
        
        #Iterating upto num1
        while idx < num1:
            #If the num can be divided by any number other than 1 and itself, it is not a prime number
            if num1 % idx == 0:
                flag = False
                break
            
            #Increment the idx value by 1 on every iteration
            idx += 1
            
        return flag
    
    #Logic of returning the next element in the sequence
    def __next__(self):
        
        #Stopping the iteration if loop reaches n
        if self._val > self._n:
            raise StopIteration('\nThe loop has been completed !!')
        
        else:
            #Returning the boolean output on whether the current value is a prime or not
            out = self.is_prime(self._val)
            
            #Incrementing tmp val by 1 on each iteration
            self._val += 1
            return out

#Handling any errors where non integers are entered
try:
    #Entering the num till where we want the sequence to iterate
    num = int(input('Enter a positive integer: '))
    
    #Handling the scenario where a negative integer/ lesser than 2
    #is entered as sequence starts by 2 and increments with 1
    if num < 2:
        raise ValueError()
    
    print('\nThe prime numbers upto {}: '.format(num))
    
    #Enumerating using the Iterable object and displaying numbers where the __next__ output returns True
    for idx, ele in enumerate(Prime(num), start=2):
        if ele == True:
            print(idx)
    
except ValueError:
    print('\nKindly check the input provided. Enter a positive integer greater than 1.')
```

**Q7** - *Define a class, Circle, which takes two arguments when defined — a sequence and a number. The idea is that the object will then return elements the defined number of times. If the number is greater than the number of elements, then the sequence repeats as necessary.*

```python
class Circle:
    
    #Initialising the variables in the class
    def __init__(self, sequence, num):
        self._sequence = sequence
        self._num = num
        
        #Creating a tmp variable to stop or continue with the iteration
        self._val = 0
    
    #Making the object into an iterable object
    def __iter__(self):
        return self
    
    #Logic to return the next element in the sequence
    def __next__(self):
        
        #Due to zero indexing nature of python -- self._val can take max value of num - 1
        if self._val >= self._num:
            raise StopIteration
            
        #Returning the output -- Indexed element from the iterable sequence    
        out = self._sequence[self._val % len(self._sequence)]
        self._val += 1
        return out

try: 
    #Enter a sequence of characters/ word
    ls = input('Enter a sequence as a continous string: ')

    #Maximum number of times to sequence the ls
    max_times = int(input('Enter the max num of times to sequence the list: '))
    print('\nThe resulting cyclic sequence "{}" with max repetitions {}:\n'.format(ls, max_times))
    
    #Defining an iterable object of the class Circle 
    iter_obj = iter(Circle(ls, max_times))
    
    for i in iter_obj:
        print(i, end = ' ')

except:
    print('Kindly check the inputs provided.')
```

**Q8** - *Write a Python program to demonstrate working of iterators using an example type that iterates from 10 to given value*

```python
#Assuming the iterators starts from 10 and can iterate in any direction by either increment/ decrement by 1
#Valid form of inputs are only integers

class pos_neg_iterator:
        
    #Initialising the variables required
    def __init__(self, n):
        self._n = n
        
        #Starting the loop from 10
        self._val = 10
        
        #Defining a flag to build bidirectional logic of iteration in __next__ constructor 
        self._flag = self._n > self._val
    
    #Making the object iterable    
    def __iter__(self):
        return self
    
    #Logic of returning the next element in the sequence -- Decreasing in val
    def __next__(self):
        
        #Writing logic to iterate on either direction from 10
        #When number provided by user is greater than 10
        if self._flag == True:
            
            #Stopping the iteration if loop reaches n
            if self._val > self._n:
                raise StopIteration('\nThe loop has been completed !!')
        
            else:
                out = self._val
            
                #Incrementing the val by 1 on each iteration
                self._val += 1
            
                #Returning the current value of the tmp variable _val
                return out
        
        else:
            #Stopping the iteration if tmp _val becomes lesser than  n
            if self._val < self._n:
                raise StopIteration('\nThe loop has been completed !!')
        
            else:
                out = self._val
            
                #Decrementing the val by 1 on each iteration
                self._val -= 1
            
                #Returning the current value of the tmp variable _val
                return out

#Handling any errors where non integers are entered
try:
    #Entering the num till where we want the sequence to iterate
    num = int(input('Enter a positive/ negative integer: '))
    
    print('\nThe loop iterates from 10 to {}: '.format(num))
    
    #Iterating over the iterable class -- Bidirectional situation handles
    for ele in pos_neg_iterator(num):
        print(ele)
    
except ValueError:
    print('\nKindly check the input provided. Enter a positive integer.')
```

**Q9** - *Write a program to illustrate the concept of multiple decorators*

```python
#Defining the decorators to be used 
def top_bottom1(func):
    
    #Defining the wrapper function
    def wrapper1(*args, **kwargs):
        
        #Upper and lower bounding the pyramid with '#' characters
        print('#' * 42)
        func(*args, **kwargs)
        print('#' * 42)
        
    return wrapper1

def top_bottom2(func):
    
    #Defining the wrapper function
    def wrapper2(*args, **kwargs):
        
        #It runs on top of the output from the top_bottom1 decorator result
        print('Start....\n')
        print('@%' * 21)
        func(*args, **kwargs)
        print('@%' * 21)
        print('\nEnd....')
        
    return wrapper2 

#Implementing factorial using aggregation
def agg_fact(num):
    """
    Takes a number as input and returns its factorial
    """
    #Creating an aggregating variable
    prod = 1
    
    #Iterating from 1 to num
    for n in range(1, num+1):
        prod *= n
        
    return prod

def invert(num):
    """
    Takes a number and inverts it along the y axis
    """
    #Converting the number into string
    st = str(num)
    
    #Using the join method on the list iterable containing reversed numbers
    return ''.join(list(reversed(st)))

#With the current spacing provided the function can work well with input num values <= 21
#Using multiple decorators that start from bottom and go to top in order of execution 

@top_bottom2
@top_bottom1
def display_inverted(num):
    """
    Takes a num as arguments and returns an inverted pyramid containing
    factorial values printed from 0 to num and num to 0
    """
    #Iterating from 0 to num
    for n1 in range(num+1):
        print('{0:20}{1:1}'.format(agg_fact(n1), invert(agg_fact(n1))))
        
        #print(agg_fact(n1), agg_fact(n1), sep = ' ', end = '\n')
    for n2 in reversed(range(num+1)):
        print('{0:20}{1:1}'.format(agg_fact(n2), invert(agg_fact(n2))))
        
        #print(agg_fact(n2) , agg_fact(n2), sep = ' ',end = '\n')

try:
    #Printing the final result
    num = int(input('Enter a positive integer less than or equal to 21: '))
    
    if num < 0:
        raise ValueError
    
    display_inverted(num)

except:
    print('Kindly check the value entored. Only positive integers <=20 allowed') 
```

**Q10** - *Write a decorator called timer that calculates the time required to execute a function. Its wrapper function called calculates should take all the parameters that are passed to the decorated function.*

```python
#Defining the timer decorator to get execution speed
def timer(func):
    
    #Creating a wrapper function calculates to add additional complexity
    def calculates(*args, **kwargs):
        
        #Starting the time counter
        start_time = time.perf_counter()
        
        #The function value is stored in the variable output
        output = func(*args, **kwargs)
        
        #Ending the time counter
        end_time = time.perf_counter()
        exe_total = end_time - start_time
        
        #Printing the execution time in milliseconds
        print('\nThe execution speed of the program: {:.3f} ms'.format(exe_total * (10**6)))
        return output
    return calculates

#Defining two programs and checking their respective execution speeds -- Using the timer decorator

@timer
def agg_fact(num):
    """
    Takes a number as input and returns its factorial
    """
    #Creating an aggregating variable
    prod = 1
    
    #Iterating from 1 to num
    for n in range(1, num+1):
        prod *= n
        
    return prod

@timer
def is_prime(num1):
    """
    Returns a boolean output on whether num1
    is prime or not
    """
    
    if num1 < 2:
        return False
    
    else:
        #Creating local temp variables for displaying result
        flag = True
        idx = 2

        #Iterating upto num1
        while idx < num1:
            #If the num can be divided by any number other than 1 and itself, it is not a prime number
            if num1 % idx == 0:
                flag = False
                break

            #Increment the idx value by 1 on every iteration
            idx += 1

        return flag

#Error handling for input
try:
    #Accepting the user defined positive integer
    num = int(input('Enter a positive integer to calculate its factorial: '))
    
    if num < 0:
        raise ValueError

    print('Whether the number {} is prime or not: {}'.format(num, is_prime(num)))
    print('\n' + '*'*20 + '\n')
    print('The result of the aggregated factorial solution: {}'.format(agg_fact(num)))
    
except:
    print('\nKindly check the input entered.')
```
